## Business Understanding

E-Commerce Public Dataset merupakan kumpulan data transaksi dari sebuah platform e-commerce yang mencakup informasi pelanggan, pesanan, pembayaran, produk, penjual, dan ulasan pelanggan.

Analisis ini bertujuan untuk memahami pola penjualan dan perilaku pelanggan sehingga dapat digunakan sebagai dasar pengambilan keputusan bisnis yang lebih efektif.

## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

Metode pembayaran apa yang memiliki rata-rata nilai transaksi tertinggi selama periode analisis?

Specific: Fokus pada metode pembayaran pelanggan.

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
sns.set_theme(style="whitegrid")
```

## Data Wrangling

### Gathering Data

#### Load df ...

```python
customers_df = pd.read_csv("data/customers_dataset.csv")
orders_df = pd.read_csv("data/orders_dataset.csv")
order_items_df = pd.read_csv("data/order_items_dataset.csv")
payments_df = pd.read_csv("data/order_payments_dataset.csv")
products_df = pd.read_csv("data/products_dataset.csv")
translation_df = pd.read_csv("data/product_category_name_translation.csv")
reviews_df = pd.read_csv("data/order_reviews_dataset.csv")
sellers_df = pd.read_csv("data/sellers_dataset.csv")
geolocation_df = pd.read_csv("data/geolocation_dataset.csv")
```

**Insight:** (Opsional)

-Dataset terdiri atas sembilan tabel utama yang saling berkaitan, meliputi informasi pelanggan, pesanan, produk, pembayaran, penjual, ulasan, serta lokasi geografis.

-Dataset dengan jumlah observasi terbesar adalah geolocation_dataset dengan lebih dari satu juta baris data.

### Assessing Data

#### Identifying ... problem
1.Missing value pada products_dataset.

2.Missing value pada orders_dataset.

3.Missing value pada order_reviews_dataset.

4.Duplicate data pada geolocation_dataset.

5.Kolom tanggal masih bertipe object.

```python
datasets = {
    "Customers": customers_df,
    "Orders": orders_df,
    "Order Items": order_items_df,
    "Payments": payments_df,
    "Products": products_df,
    "Translation": translation_df,
    "Reviews": reviews_df,
    "Sellers": sellers_df,
    "Geolocation": geolocation_df
}

for name, df in datasets.items():
    print(df.isnull().sum())
    print(df.duplicated().sum())
```

**Steps to Take:**

- Mengisi missing value kategori produk dengan nilai "unknown".

- Mengisi missing value numerik menggunakan median.

- komentar review kosong dengan placeholder.

- Menghapus duplicate pada geolocation.

- Mengubah kolom tanggal menjadi datetime.

**Insight:** (Opsional)
- Permasalahan kualitas data cukup umum ditemukan pada data transaksi berskala besar sehingga proses cleaning diperlukan agar hasil analisis lebih akurat.

### Cleaning Data

#### Fixing ... problem

```python
geolocation_df.drop_duplicates(inplace=True)

products_df["product_category_name"] = (
    products_df["product_category_name"]
    .fillna("unknown")
)
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    orders_df[col] = pd.to_datetime(
        orders_df[col],
        errors="coerce"
    )
numeric_cols = [
    "product_name_lenght",
    "product_description_lenght",
    "product_photos_qty",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

for col in numeric_cols:
    products_df[col] = products_df[col].fillna(
        products_df[col].median()
    )

reviews_df["review_comment_title"] = (
    reviews_df["review_comment_title"]
    .fillna("No Title")
)

reviews_df["review_comment_message"] = (
    reviews_df["review_comment_message"]
    .fillna("No Comment")
)

print("Missing Values Setelah Cleaning")

print("\nProducts")
print(products_df.isnull().sum())

print("\nReviews")
print(reviews_df.isnull().sum())

print("\nGeolocation Duplicate")
print(geolocation_df.duplicated().sum())
```

**Insight:** (Opsional)
- Setelah cleaning dilakukan, dataset menjadi lebih konsisten dan siap digunakan untuk tahap eksplorasi.
- Seluruh missing value pada kolom yang relevan telah ditangani dan duplicate data pada geolocation berhasil dihapus sehingga dataset siap digunakan untuk analisis.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
all_df = orders_df.merge(
    payments_df,
    on="order_id",
    how="left"
)

all_df = all_df.merge(
    order_items_df,
    on="order_id",
    how="left"
)

all_df = all_df.merge(
    products_df,
    on="product_id",
    how="left"
)

all_df = all_df.merge(
    translation_df,
    on="product_category_name",
    how="left"
)

all_df = all_df.merge(
    customers_df,
    on="customer_id",
    how="left"
)

# Revenue bulanan
revenue_df = orders_df.merge(
    payments_df,
    on="order_id",
    how="left"
)

revenue_df["year_month"] = (
    revenue_df["order_purchase_timestamp"]
    .dt.to_period("M")
    .astype(str)
)

monthly_revenue = (
    revenue_df.groupby("year_month")["payment_value"]
    .sum()
    .reset_index()
)

# Revenue kategori produk
category_df = (
    order_items_df
    .merge(products_df, on="product_id", how="left")
    .merge(
        translation_df,
        on="product_category_name",
        how="left"
    )
)

category_df["item_revenue"] = (
    category_df["price"] +
    category_df["freight_value"]
)

category_revenue = (
    category_df.groupby("product_category_name_english")
    ["item_revenue"]
    .sum()
    .sort_values(ascending=False)
)

# Rata-rata transaksi per metode pembayaran
payment_average = (
    payments_df.groupby("payment_type")["payment_value"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

print(all_df.shape)

all_df.head()
print("Shape:", all_df.shape)

print("\nMissing Values:")
print(all_df.isnull().sum().sort_values(ascending=False).head(10))

print("\nData Types:")
print(all_df.dtypes)

all_df.head()
```

**Insight:** (Opsional)
- Revenue menunjukkan tren meningkat dari tahun 2016 hingga pertengahan 2018.

- Kategori produk tertentu memberikan kontribusi revenue yang jauh lebih besar dibanding kategori lainnya.

- Metode pembayaran menunjukkan karakteristik nilai transaksi yang berbeda.

- Dataset hasil penggabungan terdiri dari informasi transaksi, pembayaran, pelanggan, dan kategori produk yang akan digunakan untuk menjawab pertanyaan bisnis.

## Visualization & Explanatory Analysis

### Pertanyaan 1:
Bagaimana tren revenue bulanan?

```python
plt.figure(figsize=(14,6))

plt.plot(
    monthly_revenue["year_month"],
    monthly_revenue["payment_value"],
    marker="o"
)

plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:
Kategori produk apa yang menghasilkan revenue tertinggi?



```python
top_10_categories = (
    category_revenue
    .reset_index()
    .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=top_10_categories,
    x="item_revenue",
    y="product_category_name_english"
)

plt.title("Top 10 Product Categories by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Product Category")

plt.tight_layout()
plt.show()
```

### Pertanyaan 3:
Metode pembayaran apa yang memiliki rata-rata transaksi tertinggi?

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=payment_average,
    x="payment_type",
    y="payment_value"
)

plt.title("Average Transaction Value by Payment Type")
plt.xlabel("Payment Type")
plt.ylabel("Average Transaction Value")

plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)
- Revenue mengalami peningkatan bertahap sepanjang tahun 2017 dan mencapai puncaknya pada November 2017 dengan total revenue sekitar 1,2 juta. Setelah itu revenue tetap tinggi hingga pertengahan 2018 sebelum menurun pada akhir periode dataset.
- Kategori bed_bath_table merupakan penyumbang revenue terbesar, diikuti oleh health_beauty dan computers_accessories. Ketiga kategori tersebut memiliki potensi besar untuk diprioritaskan dalam strategi bisnis.
- Metode pembayaran credit_card memiliki rata-rata nilai transaksi tertinggi dibanding metode lainnya, menunjukkan bahwa pelanggan pengguna kartu kredit cenderung melakukan pembelian dengan nominal lebih besar.
